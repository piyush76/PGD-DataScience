* Education - in no. of years, of the customer.
* Marital Status - of the customer.
* Usage - Avg. # times the customer wants to use the treadmill every week.
* Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit).
* Income - of the customer.
* Miles- expected to run.

# <span style="font-family: Arial; font-weight:bold;font-size:.9em;color:#0e92ea"> Multivariate Analysis of Cardio Good Fitness Data

```python
#Product and Age relation with gender
sns.boxplot(x="Product", y="Age", data=fitness_data, hue='Gender')
plt.show()
```

```python
#Average Fitness Age of Customer per Product 
sns.barplot(data =fitness_data.groupby(by=['Product'])['Age'].mean().reset_index().sort_values(['Age']).tail(10), x="Product", y="Age")
plt.show()
```

```python
#Product and Education relation with gender
sns.boxplot(x="Product", y="Education", data=fitness_data, hue='Gender')
plt.show()
```

```python
#The Bar chart between the product and years of education for the customers.
sns.barplot(data =fitness_data.groupby(by=['Product'])['Education'].mean().reset_index().sort_values(['Education']).tail(10), x="Product", y="Education")
plt.show()
```

```python
#Marital status data analysis with product
sns.boxplot(x="MaritalStatus", y="Income", data=fitness_data, hue='Product')
plt.show()
```

```python
## Product distribution with the marital status and income data
sns.boxplot(x="Product", y="Income", data=fitness_data, hue='MaritalStatus')
plt.show()
```

```python
## Product distribution with the marital status and Usuage data
sns.boxplot(x="Product", y="Usage", data=fitness_data, hue='MaritalStatus')
plt.show()
```

```python
#Gender and Income distribution
sns.boxplot(x="Product", y="Income", data=fitness_data, hue='Gender')
plt.show()
```

```python
#Average Income data for customer per product 
sns.barplot(data =fitness_data.groupby(by=['Product'])['Income'].mean().reset_index().sort_values(['Income']).tail(10), x="Product", y="Income")
plt.show()

```

```python
#Average Usage per machine 
sns.barplot(data =fitness_data.groupby(by=['Product'])['Usage'].mean().reset_index().sort_values(['Usage']).tail(10), x="Product", y="Usage" )
plt.show()
```

```python
#Average Fitness of Customer per machine 
sns.barplot(data =fitness_data.groupby(by=['Product'])['Fitness'].mean().reset_index().sort_values(['Fitness']).tail(10), x="Product", y="Fitness" )
plt.show()
```

```python
#Average Miles per machine 
sns.barplot(data =fitness_data.groupby(by=['Product'])['Miles'].mean().reset_index().sort_values(['Miles']).tail(10), x="Product", y="Miles" )
plt.show()
```
