# Tips Dataset


`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

### Q1] Import the libraries

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

### Q3] Understand the distribution of 'tip' and 'total_bill' using different plots

Hint: Use distplot and boxplot

```python
sns.displot(tips_data.total_bill )
```

```python
sns.displot(tips_data.tip )
```

```python
sns.distplot(tips_data.tip, color='red')
plt.show()
sns.distplot(tips_data.total_bill, color='hotpink')
plt.show()
```

```python
sns.boxplot(data = tips_data, x ='total_bill');
plt.show()
sns.boxplot(data= tips_data, x='tip');
plt.show()
```

### Q4] Relationship between 'total_bill' and 'tip'

Implement jointplot, hex jointplot, pairplot and lmplot

```python
sns.jointplot(tips_data['total_bill'], tips_data['tip'],color='green')
plt.show()
sns.jointplot(tips_data['total_bill'], tips_data['tip'],kind='hex' )
plt.show()


```

```python
sns.pairplot(tips_data[['total_bill','tip']])
plt.show()

sns.lmplot(data = tips_data, x='total_bill' , y='tip')
plt.show()
```

### Q5] Make plots to check distribution of tip with different gender. 

Use striplot and boxplot

```python
sns.stripplot(x=tips_data['sex'], y=tips_data['tip'])
plt.show()
```

```python
sns.boxplot(x=tips_data['sex'], y=tips_data['tip'])
plt.show()
```

### Q6] Draw relationship between 'total_bill', 'tip' and 'sex'

use lmplot

```python
sns.lmplot(x= 'total_bill',y='tip', data=tips_data)
plt.show

##IF YOU WANT TO SHOW THE RELATIONSHIP 
sns.lmplot(x='total_bill',y='tip',data=tips_data,fit_reg=False,hue='sex')
plt.show()

sns.lmplot(x='total_bill',y='tip',data=tips_data,fit_reg=True,hue='sex')
plt.show()

```

### Q7] Which day bring in higher bill amount?

```python
sns.stripplot(x=tips_data['day'], y=tips_data['total_bill'] ,data=tips_data, jitter=False )
plt.show()
```

### Q8] Find out count of total males and females visiting each day.

use countplot

```python
sns.countplot(x=tips_data['day'], hue= tips_data['sex'] )
plt.show()

sns.countplot(x=tips_data['size'], hue= tips_data['day'] )
plt.show()
```

### Q9] At what time of which day is the total bill highest with respect to tips (consider for male and female separately)

use lmplot

```python
sns.lmplot(x='total_bill',y='tip',data=tips_data,fit_reg=False,hue='day',row='sex',col='time')
plt.show()
```
